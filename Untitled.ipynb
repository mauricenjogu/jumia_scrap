```python
item_list= []
try:
    for item in items:
        name = item.find('h3',attrs={"class":'name'}).text
        price = item.find('div',attrs={"class":'prc'}).text
        old_price_element = item.find('div',attrs={"class":'old'})
        discount_element = item.find('div',attrs={"class":'bdg _dsct _sm'})
        link = item.find('a',attrs={"class":'core'}).get('href')
        if old_price_element:
            old_price = old_price_element.text
        else:
            old_price = None
        if discount_element:
            discount = discount_element.text
        else:
            discount = None
        item_list.append((name,price,old_price,discount,link))
except AttributeError:
    item_list.append(None)
item_list
```

```python
item_list= []
import time
base_url = 'https://www.jumia.co.ke/all-products/?page={}#catalog-listing'
for page_num in range(1,51):
    url = base_url.format(page_num)
    time.sleep(5)
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        print(f'page {page_num} scraped successfully')
    else:
        print(f'page {page_num} failed')
    items = soup.find_all('article', attrs={"class":'prd _fb col c-prd'})
    

    for item in items:
        try:
            name = item.find('h3',attrs={"class":'name'}).text
            price = item.find('div',attrs={"class":'prc'}).text
            old_price_element = item.find('div',attrs={"class":'old'})
            discount_element = item.find('div',attrs={"class":'bdg _dsct _sm'})
            link = item.find('a',attrs={"class":'core'}).get('href')
            if old_price_element:
                old_price = old_price_element.text
            else:
                old_price = None
            if discount_element:
                discount = discount_element.text
            else:
                discount = None
            item_list.append((name,price,old_price,discount,link))
        except AttributeError:
            item_list.append(None)
```

```python
len(item_list)
```

```python
df=pd.DataFrame(item_list,columns=['name','price','old_price','discount','link'])
df
```

```python
df.info()
```

```python
df['price'] = df['price'].apply(lambda x: x[4:] )
```

```python
df = df.dropna()
```

```python
df['old_price'] = df['old_price'].apply(lambda x: x[4:] )
df['discount'] = df['discount'].apply(lambda x: x[:2] )
df
```

```python
selected_rows = df[df['price'].str.contains('-')| df['old_price'].str.contains('-')]

selected_rows
```

```python
# Assuming df is your DataFrame and 'column_name' is the column to split
split_values = selected_rows['price'].str.split('-', expand=True)

# Extract start_value and end_value from split_values
selected_rows.loc[:, 'price_start_value'] = split_values[0].str.strip()
selected_rows.loc[:, 'price_end_value'] = split_values[1].str.strip()

# Fill missing end_value with start_value
selected_rows['price_end_value'].fillna(selected_rows['price_start_value'], inplace=True)

selected_rows
```

```python
# Assuming df is your DataFrame and 'column_name' is the column to split
split_values = selected_rows['old_price'].str.split('-', expand=True)

# Extract start_value and end_value from split_values
selected_rows.loc[:, 'old_price_start_value'] = split_values[0].str.strip()
selected_rows.loc[:, 'old_price_end_value'] = split_values[1].str.strip()

# Fill missing end_value with start_value
selected_rows['old_price_end_value'].fillna(selected_rows['old_price_start_value'], inplace=True)

selected_rows
```

```python
selected_rows.drop(['price','old_price'],axis = 1,inplace=True)
```

```python
columns_to_convert = ['price_start_value', 'price_end_value', 'old_price_start_value', 'old_price_end_value']

# Function to convert string values to float, handling 'KSh' and commas
def convert_to_float(value):
    if isinstance(value, str):
        if 'KSh' in value:
            value = value.replace('KSh', '').replace(',', '').strip()
        else:
            value = value.replace(',', '').strip()
    return pd.to_numeric(value, errors='coerce')

# Convert values in specified columns to float
for column in columns_to_convert:
    selected_rows[column] = selected_rows[column].apply(convert_to_float)
selected_rows
```

```python
selected_rows['price']=(selected_rows['price_start_value'] + selected_rows['price_end_value'])/2
selected_rows['old_price']=(selected_rows['old_price_start_value'] + selected_rows['old_price_end_value'])/2
selected_rows.head(2)
```

```python
selected_rows.drop(['price_start_value','price_end_value','old_price_start_value','old_price_end_value'],axis=1,inplace= True)
selected_rows.head()
```
